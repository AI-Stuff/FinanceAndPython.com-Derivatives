```python
import matplotlib.pyplot as plt
stockPrices = [50,70,90,110,130]
profit1 = [x-100 for x in stockPrices]
profit2 = [max(90-x,0) for x in stockPrices]
plt.plot(stockPrices,profit1)
plt.plot(stockPrices,profit2)
plt.legend(["Profit on Stock","Profit on Put"])

plt.xlabel("Stock Prices")
plt.ylabel("Profit")
plt.title("Insuring a Stock Position")
plt.show()
```
