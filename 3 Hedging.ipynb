```python
import matplotlib.pyplot as plt
def stockReturn(S1,S0):
    return (S1-S0)/S0
prices = list(range(100,201))
stockReturns = [stockReturn(x,150) for x in prices]
plt.plot(prices,stockReturns,"k")
plt.axvline(150, linestyle='dashed', linewidth=1)
plt.axhline(0, linestyle='dashed', linewidth=1)
plt.xlabel("Stock Price in Period 1")
plt.ylabel("Return")
plt.title("Classic Stock Payoff")
plt.show()
```

```python
def stockReturn(S1,S0):
    return ((S1-S0)-2+max(0,135-S1))/S0
prices = list(range(100,201))
stockReturns = [stockReturn(x,150) for x in prices]
plt.plot(prices,stockReturns,"k")
plt.axvline(150, linestyle='dashed', linewidth=1)
plt.axhline(0, linestyle='dashed', linewidth=1)
plt.xlabel("Stock Price in Period 1")
plt.ylabel("Return")
plt.title("Protective Put")
plt.show()
```

```python
def stockReturn(S1,S0):
    return ((S1-S0)-2+max(0,135-S1)+2-max(0,S1-165))/S0
prices = list(range(100,201))
stockReturns = [stockReturn(x,150) for x in prices]
plt.plot(prices,stockReturns,"k")
plt.axvline(150, linestyle='dashed', linewidth=1)
plt.axhline(0, linestyle='dashed', linewidth=1)
plt.xlabel("Stock Price in Period 1")
plt.ylabel("Return")
plt.title("Collar Spread")
plt.show()
```

```python
np.random.seed((1))
points = np.random.normal(.05,.08,100000)
```

```python
import matplotlib.pyplot as plt
plt.hist(points, 30)
plt.show()
```

```python
prices = [(x+1)*150 for x in points]
returns1 = [((x-150)-2+max(0,135-x))/150 for x in prices]
returns1 = np.array(returns1)
plt.hist(returns1, 30)
plt.show()
```

```python
prices = [(x+1)*150 for x in points]
returns2 = [((x-150)-2+max(0,135-x)+2-max(0,x-180))/150 for x in prices]
returns2 = np.array(returns2)
plt.hist(returns2, 30)
plt.show()
```
