```python
def forwardArbitrage(X,s,r,t):
    #First see if they are equal, in which case there is no arbitrage
    if X/(1+r)**t == s:
        print("There is no arbitrage opportunities")
    #If the forward contract is less than we want to buy it, and sell the underlying asset
    elif X/(1+r)**t < s:
        print("In period 0.....")
        print("We buy the forward contract, costing us $0")
        print("We short the stock, which brings in $"+str(s))
        print("We invest all these proceeds at the market rate, thus we have $0 left over.")
        print("Our inflows in period 0 is $0")
        print("")
        print("In period 1....")
        print("Our forward contract expires, so we buy the stock for $"+str(X))
        print("We immediately use this stock to end our short sale of the stock from period 0.")
        print("Finally, we get our invested money back with a value of $"+str(s*(1+r)**t))
        print("The total cash inflow in this period for us is: $"+str(s*(1+r)**t-X))
    #Otherwise, do the opposite
    else:
        print("In period 0.....")
        print("We sell the forward contract, costing us $0")
        print("We borrow money at the market rate totalling: $"+str(s))
        print("We buy the stock, which costs $"+str(s))
        print("Our inflows in period 0 are $0")
        print("")
        print("In period 1....")
        print("Our forward contract expires, so we sell the stock we bought in period 1 for $"+str(X))
        print("We also pay back our loan which costs us $"+str(s*(1+r)**t))
        print("The total cash inflow in this period for us is: $"+str(X-s*(1+r)**t))
```

```python
forwardArbitrage(105,100,.05,1)
```

```python
forwardArbitrage(102,100,.05,1)
```

```python
import matplotlib.pyplot as plt
r = .05
t = 1
X = 105
prices = list(range(70,130))
profit = []
for price in prices:
    if X/(1+r)**t == price:
        profit.append(0)
    elif X/(1+r)**t < price:
        profit.append(price*(1+r)**t-X)
    else:
        profit.append(X-price*(1+r)**t)
plt.plot(prices,profit)
plt.xlabel("Stock Price")
plt.ylabel("Profit")
plt.title("Forward Contract Arbitrage")
plt.show()
```
